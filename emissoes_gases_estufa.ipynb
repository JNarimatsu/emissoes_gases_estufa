### Transformando as colunas de anos em uma única coluna Ano

```python
colunas_info = list(emissoes_gases.loc[:, 'Nível 1 - Setor': 'Produto'].columns)
colunas_info
```

```python
colunas_emissao = list(emissoes_gases.loc[:, 1970:2021].columns)
colunas_emissao
```

```python
#Utilizamos o método melt pois com ele é possivel transformar um dataframe de um formato amplo(wide) para o formato longo(long)
emissoes_por_ano = emissoes_gases.melt(id_vars= colunas_info, value_vars=colunas_emissao, var_name = 'Ano', value_name = 'Emissão')
```

```python
emissoes_por_ano
```

## Agrupando por tipo de gás

```python
emissoes_por_ano.groupby('Gás').groups
```

```python
emissoes_por_ano.groupby('Gás').get_group('CO2 (t)')
```

## Soma total de emissão por tipo de gás

```python
emissoes_por_ano.groupby('Gás').sum(numeric_only = True)
```

```python
emissoes_por_ano.groupby('Gás')[['Emissão']].sum()
#Essa forma é aconselhada caso tenha mais colunas numericas
```

```python
emissoes_por_gas = emissoes_por_ano.groupby('Gás')[['Emissão']].sum().sort_values('Emissão', ascending = False)
emissoes_por_gas
```

```python
emissoes_por_gas.plot(kind = 'barh', figsize = (10,6), color = 'wheat');
```

```python
#Obtendo a proporção de CO2 em relação aos demais gases
emissoes_por_gas.iloc[0:9]
```

```python
print(f'A emissao de CO2 corresponde a {float((emissoes_por_gas.iloc[0:9].sum()/emissoes_por_gas.sum()).iloc[0])*100: .2f} % de emissssão total de gases estufa no Brasil de 1970 a 2021.')
```
